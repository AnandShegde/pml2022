```python
import jax.numpy as jnp
import numpy as np
import jax
import matplotlib.pyplot as plt
import math
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def f(x):
    return jnp.cos(x)
```

```python
x_range = jnp.linspace(-5, 5, 1000)
plt.plot(x_range, f(x_range))
```

```python
### Oth order approx

g = {}
g[0] = jax.vmap(lambda x: 1)
plt.plot(x_range, f(x_range), label='True')
plt.plot(x_range, g[0](x_range), label='0th order approximation')
plt.legend()
```

```python
order_approx_max = 20
grads = {0: f}
for i in range(1, order_approx_max):
    grads[i] = jax.grad(grads[i-1])
```

```python
for i in range(0, 4):
    plt.plot(x_range, jax.vmap(grads[i])(x_range), label='f'+"'"*i)
plt.legend()
```

```python
import time
out = {}
for i in range(1, 12):
    start = time.time()
    grads[i](0.)
    end = time.time()
    out[i] = end - start

print(out)
```

```python
order_approx = 11
coeffs = [None]*order_approx
coeffs[0] = f(0).item()
for i in range(1, order_approx):
    coeffs[i] = grads[i](0.).item()/math.factorial(i)
```

```python
coeffs
```

```python
def appx(coeffs):

    g = lambda x: sum([coef*(x**i) for i, coef in enumerate(coeffs)])
    return g
```

```python
plt.plot(x_range, f(x_range), label='True')

plt.plot(x_range, appx(coeffs)(x_range), label='Appx. Polynomial')
plt.legend()
```
